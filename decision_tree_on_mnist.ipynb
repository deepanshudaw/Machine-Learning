```
from sklearn import tree
```

```
import tensorflow as tf
mnist = tf.keras.datasets.mnist
(train_images, ytrain), (test_images, ytest) = mnist.load_data()
train_images = train_images / 255.0
test_images = test_images / 255.0
xtrain = train_images.reshape(60000,784)
xtest = test_images.reshape(10000,784)
```

```
clf=tree.DecisionTreeClassifier()
clf.fit(xtrain,ytrain)
```

```
expected=(ytest)
accuracy=0
predicted=clf.predict(xtest)
for i in range(10000):
  if (expected[i]==predicted[i]):
    accuracy=accuracy+1
print("Prediction Accuracy:",accuracy/100)
```

I used the code given below(copied) to display the tree

```
from sklearn.externals.six import StringIO  
from IPython.display import Image  
from sklearn.tree import export_graphviz
import pydotplus
dot_data = StringIO()
export_graphviz(clf, out_file=dot_data,  
                filled=True, rounded=True,
                special_characters=True)
graph = pydotplus.graph_from_dot_data(dot_data.getvalue())  
Image(graph.create_png())
```
